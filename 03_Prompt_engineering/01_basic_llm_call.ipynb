```python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

client = OpenAI()
```

```python
def call_llm(prompt):
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )
    print(response.choices[0].message.content)
```

```python
call_llm("classify the email : My invoice is wrong, please fix it.")
```

```python
# emails

def call_llm(prompt):
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    return response.choices[0].message.content
```

```python
email = emails[25]['body']
```

```python
prompt_v2 = f"""You are an expert support email classifier. 
Classify the email into exactly one of: 
[Billing, Technical, Feature Request, Spam, Other]. 
Return only the category name,
nothing else. Email: {email}"""

call_llm(prompt_v2)
```

```python
# basic prompt
for i in range(5):
    email =emails[i]['body']
    prompt = f"classify the below email : {email}"
    output = call_llm(prompt)
    emails[i]['model_output'] = output
    # print('output:', output)
```

```python

for i in range(10):
    email = emails[i]['body']
    subject = emails[i]['subject']
    prompt = f"""You are an expert support email classifier. 
Classify the email into exactly one of category: 
Return only the category name,
nothing else. Email: {email}
Subject : {subject}
"""
    output = call_llm(prompt)
    emails[i]['model_output'] = output
```
