```python
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

client = OpenAI()
```

```python
def call_llm(prompt):
    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ]
    )
    print(response.choices[0].message.content)
```

```python
call_llm("classify the email : My invoice is wrong, please fix it.")
```
